```python
import cv2
import PIL
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
model = YOLO("../mini_project/model/yolo11/Seg/yolo11n-seg.pt")
img = "https://ultralytics.com/images/bus.jpg"
results = model.predict(source=img,conf=0.88,retina_masks=True)
```

```python
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.save(filename="../mini_project/tmp/result.jpg")  # save to disk
```

```python

plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
results[0].save(filename="../mini_project/tmp/result.jpg")
```
